```python
def normalize(s):
    l = [
        ['a', 'á'],
        ['e', 'é'],
        ['i', 'í'],
        ['o', 'ó'],
        ['u', 'ú']
    ]
    
    for i in l:
        s.replace(i[1], i[0])
        
    return s
```

```python
URL = 'https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses#Estados_asociados_a_estados_soberanos'
page = requests.get(URL)

content = page.content

soup = BeautifulSoup(content, 'html.parser')
table = soup.find_all(class_ = "wikitable sortable col1cen")[0].find_all('tbody')[0].find_all('tr')[1:]

country_list = []
count = 1
for i in table:
    
    country = i.find_all('td')[1].text
    country = country.strip().lower()
    
    capital = ''
    if i.find_all('td')[3].a is None :
        capital = i.find_all('td')[3].text
    else:
        capital =i.find_all('td')[3].a.text 
        
    capital = capital.strip().lower()
    
    country_link = i.find_all('td')[0].find_all('a')[1]['href']
    wiki = "https://es.wikipedia.org" + country_link
    
    
    country_list.append([
        normalize(country), 
        normalize(capital), 
        normalize(wiki)
    ])
    
    print("pais N°: " + str(count))
    print("nombre - capital: " + str(normalize(country)) + ' - ' + str(normalize(capital)))
    
    count = count + 1

print("total: " + str(count - 1))
```

```python
final_country_list = []
for i in country_list:
    wiki = i[2]
    page1 = requests.get(wiki)
    soup1 = BeautifulSoup(page1.content, 'html.parser')
    
    tabla = soup1.find_all('table', class_ = "infobox geography vcard")[0].tbody.contents
    
    idioma = ''
    gentilicio = ''
    country = i[0]
    capital = i[1]
    print("country, capital: " + country + " - " + capital)
    c = 0
    for j in tabla:
        c = c + 1
        
        if len(j.contents) < 2:
            continue
        
        if j.contents[0].a is None :
            continue
            
        if 'idioma' in (j.contents[0].a.text).lower():
            idioma = j.contents[1].text.lower().strip()
            print('idioma: ' + idioma)
            
        if 'gentilicio' in (j.contents[0].a.text).lower():
            gentilicio = j.contents[1].text.lower().strip()
            
            for id in range(len(gentilicio)):
                if not gentilicio[id].isalpha():
                    gentilicio = gentilicio[0:id]
                    break
            
            print('gentilicio: ' + gentilicio + "\n")
            
            
    final_country_list.append([
        normalize(country), 
        normalize(capital),
        normalize(idioma), 
        normalize(gentilicio)])
    
```

```python
# Extraccion de los verbos. 
path = "D:\\Documents\\Memoria - Eval. Word Embeddings\\Data\\SimilarBATS_español\\Inflexion\\_word_list_verbs.txt"

verbs = []

with io.open(path, 'r') as f:
    for l in f:
        verbs.append(l.strip())
        
verbs = list(dict.fromkeys(verbs))
print("cantidad de verbos: " + str(len(verbs)))
print(verbs)
```

```python
# Definir conjugaciones a usar
persona = ["1_sing", "2_sing_tu", "2_sing_usted", "3_sing", "1_plur", "2_plur_vosotros", "2_plur_ustedes", "3_plur"]
tiempo = ["presente", "preterito_imp", "presente_perf", "futuro_imp", "cond_simp"]

nombre_tiempos = []
for t in tiempo:
    for p in persona:
        nombre_tiempos.append(t + "__" + p)

nombre_conj = ["inf", "gerund", "parti"] + nombre_tiempos

print("Cantidad de conjugaciones a registrar: " + str(len(nombre_conj)))
print("Cantidad de verbos a registrar: " + str(len(verbs)))

conjugaciones = [] 
errores = 0
```

```python
# Obtencion de conjugaciones
count = 1
save = []
for s in verbs:
    print("verbo " + str(count) + ": " + s)
    count = count + 1
    
    # Obtencion de pagina desde la RAE
    url = "https://dle.rae.es/" + s
    page = requests.get(url)
    content = page.content
    soup = BeautifulSoup(content, 'html.parser')
    
    try:
        # Obtencion de formas verbales: infinitivo, gerundio y participio
        conjugacion = soup.body.find_all(id='conjugacion')[0]
        table = conjugacion.find_all('table')[0]
        tr_list = table.find_all('tr')
        
        infinitivo = tr_list[2].find_all('td')[3].text
        gerundio = tr_list[2].find_all('td')[4].text
        participio = tr_list[4].find_all('td')[3].text
        
        save = [
            infinitivo,
            gerundio,
            participio
        ]
        
        print("infinitivo: " + infinitivo)
        print("gerundio: " + gerundio)
        print("participio: " + participio + "\n")
        
        
        # Obtencion de vormas verbales para diversos tiempos
        presente = []
        preterito_imp = []
        presente_perf = []
        futuro_imp = []
        cond_simp = []
        
        for i in range(7, 15):
            s = tr_list[i].find_all('td')[3].text
            presente.append(s)
            save.append(s.strip())
            
        for i in range(7, 15):
            s = tr_list[i].find_all('td')[4].text
            preterito_imp.append(s)
            save.append(s.strip())
            
        for i in range(16, 24):
            s = tr_list[i].find_all('td')[3].text
            presente_perf.append(s)
            save.append(s.strip())
            
        for i in range(16, 24):
            s = tr_list[i].find_all('td')[4].text
            futuro_imp.append(s)
            save.append(s.strip())
            
        for i in range(25, 33):
            s = tr_list[i].find_all('td')[3].text
            cond_simp.append(s)
            save.append(s.strip())
            
        print("presente: ")
        print(presente)
        print("preterito_imp: ")
        print(preterito_imp)
        print("presente_perf: ")
        print(presente_perf)
        print("futuro_imp: ")
        print(futuro_imp)
        print("cond_simp: ")
        print(cond_simp)
        print("cantidad en save: " + str(len(save)))
            
        conjugaciones.append(save)
        print('')
        
        
    except:
        conjugaciones.append('Error: ' + s)
        errores = errores + 1
        print("ERROR")
```

```python
# Guardar verbos obtenidos.
path = "D:\\Documents\\Memoria - Eval. Word Embeddings\\Data\\SimilarBATS_español\\Inflexion\\verbs_conjugacion_final.txt"
with io.open(path, 'w') as f:
    #print(nombre_conj)
    for name in nombre_conj:
        f.write(name + '\t')
        
    f.write('\n')
    
    for verb in conjugaciones:
        #print(verb)
        for c in verb:
            v = c.strip().split()
            v = ''.join(v)
            f.write(v + '\t')
        
        f.write('\n')
        
print("Conjugaciones guardadas")
```

```python
# Generar tuplas y relaciones aleatorias
path = "D:\\Documents\\Memoria - Eval. Word Embeddings\\Data\\SimilarBATS_español\\Inflexion\\verbos_conjugacion.txt"
folder_path = "D:\\Documents\\Memoria - Eval. Word Embeddings\\Data\\SimilarBATS_español\\Inflexion" 

nombre_conjugaciones = []
verbos_conjugaciones = []
with io.open(path, 'r') as f:
    for line in f:
        if len(nombre_conjugaciones) == 0:
            nombre_conjugaciones = line.strip().split()
            
        else:
            verbos_conjugaciones.append(line.strip().split())
            
n = len(verbos_conjugaciones)
print("cantidad de verbos: " + str(n))

cont = 0
for i in range(len(nombre_conjugaciones)):
    for j in range(len(nombre_conjugaciones)):
        if j <= i:
            continue
            
        file_name = "_español_L00 [" + nombre_conjugaciones[i] + " - " + nombre_conjugaciones[j] + "].txt"
        
        id = []
        while True:
            if len(id) == 50:
                break
                
            x = random.randint(0, n - 1)
            if x in id:
                continue
                
            id.append(x)
            
        id.sort()
        #print(file_name)
        #print(id)
        
        with io.open(folder_path + "\\" + file_name, 'w') as f:
            for x in id:
                pair = verbos_conjugaciones[x][i] + "\t" + verbos_conjugaciones[x][j]
                #print(pair)
                f.write(pair + "\n")
        
        cont += 1
        
print("Count: " + str(cont))
```

```python
# Obtencion de paginas
paginas = []
path = "D:\\Documents\\Memoria - Eval. Word Embeddings\\Data\\SimilarBATS_español\\Derivacion\\paginas_derivacion.txt"

with io.open(path, 'r') as f:
    for line in f:
        paginas.append(line.strip().split())
        print(line.strip().split())
```

```python
# Obtencion de documento

total_link_palabras = []

for i in range(len(paginas)):
    time.sleep(random.randint(3, 5))
    url = paginas[i][0]
    adicion_derivacion = paginas[i][1]
    tipo_derivacion = paginas[i][2]
    
    print("==============")
    print("url: " + url)
    print("adicion_derivacion: " + adicion_derivacion)
    print("tipo_derivacion: " + tipo_derivacion)
    print("==============")
    
    page = requests.get(url)
    content = page.content
    soup = BeautifulSoup(content, 'html.parser')
    
    #print(soup.prettify())
    
    ul_list = soup.find_all('ul')
    print("len ul_list: " + str(len(ul_list)) + "\n")
    
    link_palabras = [["NEW_DOCUMENT", "", adicion_derivacion, tipo_derivacion]]
    
    for ul in ul_list:
        print("NUEVO DOC")
        li_list = ul.find_all('li')
        print("Cantidad de elementos li: " + str(len(li_list)))
        print('')
        
        
        for li in li_list:
            try:
                link_derivacion = li.find('a')['href'].strip()
                palabra_derivada = li.find('a').text.strip()
                
                link_palabras.append([link_derivacion, palabra_derivada, adicion_derivacion, tipo_derivacion])
                print(link_derivacion + " - " + palabra_derivada)
            
                
            except:
                print("ERROR")
            
        print('')
        link_palabras.append(["NEW_DOCUMENT", "", adicion_derivacion, tipo_derivacion])
        
    total_link_palabras.append(link_palabras)
```

```python
# Busqueda de las palabras.
wiki_link = "https://es.wiktionary.org"
derivacion_path = "D:\\Documents\\Memoria - Eval. Word Embeddings\\Data\\SimilarBATS_español\\Derivacion\\"

count = 0
for i in range(len(total_link_palabras)):
    save_path = derivacion_path
    save_pairs = []
    for link_y_derivacion in total_link_palabras[i]:
        link = link_y_derivacion[0]
        palabra_derivada = link_y_derivacion[1]
        adicion_derivacion = link_y_derivacion[2]
        tipo_derivacion = link_y_derivacion[3]
        
        save_path = derivacion_path + tipo_derivacion + "_" + adicion_derivacion + str(i) + ".txt"
        
        print("New link")
        print(link_y_derivacion)
        if link == "NEW_DOCUMENT":
            print('')
            
        else:
            palabra_link = wiki_link + link
            
            try:
                time.sleep(random.randint(1, 4))
                palabra_page = requests.get(palabra_link)
                palabra_content = palabra_page.content
                palabra_soup = BeautifulSoup(palabra_content, 'html.parser')
                
            except:
                print("ERROR EN CONECCION")
                continue
            
            try:
                for p in palabra_soup.find_all('p'):
                    if tipo_derivacion == "prefijo":
                        if tipo_derivacion + " " + adicion_derivacion + " y " in p.text:
                            print("buscando en: " + palabra_link)
                            p_text = p.text.strip()
                            print("texto: " + p_text)
                            
                            start = (p_text).find(adicion_derivacion + " y ") + len(adicion_derivacion + " y ")
                            end = start
                            while end < len(p_text):
                                if p_text[end] == ' ':
                                    break
                                    
                                end = end + 1
                            
                            if end == len(p_text):
                                end = -1
                                
                            original = p_text[start:end].replace(',','').replace('.', '').strip()
                            print("par de palabras: " + original + " - " + palabra_derivada + "\n")
                            save_pairs.append(original + "\t" + palabra_derivada)
                            
                            count = count + 1
                            
                    elif tipo_derivacion == "sufijo":
                        if "y el " + tipo_derivacion + " " + adicion_derivacion in p.text:
                            print("buscando en: " + palabra_link)
                            p_text = p.text.strip()
                            print("texto: " + p_text)
                            
                            start = 3
                            end = start
                            while end < len(p_text):
                                if p_text[end] == ' ':
                                    break
                                    
                                end = end + 1
                                
                            if end == len(p_text):
                                end = -1
                                
                            original = p_text[start:end].replace(',','').replace('.', '').strip()
                            print("par de palabras: " + original + " - " + palabra_derivada + "\n")
                            save_pairs.append(original + "\t" + palabra_derivada)
                            
                            count = count + 1
            except:
                print("ERROR")
                
    print(count)
    print("Guardando en " + save_path)
    with io.open(save_path, 'w') as f:
        for pair in save_pairs:
            f.write(pair + "\n")
        
```
